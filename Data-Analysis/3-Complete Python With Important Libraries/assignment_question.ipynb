**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
sum = 0
n = 100

for i in range(1, n+1):
    sum += i

print(sum)
```

**Question 18:** Write a Python program to reverse a string.

```python
str = "Your code here"
n = len(str)
reversed = ""

for i in range(0, n):
    reversed += str[n-1-i]

print(reversed)
```
