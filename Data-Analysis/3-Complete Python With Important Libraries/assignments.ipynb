### Assignment 14: Prime Number Check

Write a program that checks if a number input by the user is a prime number using a for loop.

```python
#Assignment 14 

num = int(input("Please insert a number greater than 1 to determine if is prime or not: "))

i = 2

while i < num:
    if num % i == 0:
        print("The number is not a prime number")
        break
    i = i + 1

if i == num:
    print("The number is a primer number")
```

### Assignment 13: Sum of Digits

Write a program that calculates the sum of the digits of a number input by the user using a while loop.

```python
num = int(input("Please insert a number for calculating the sum of its digits"))

aux = num
sum = 0

while aux > 0:
    sum += aux % 10
    print(aux % 10)

    aux = aux // 10 
    

print("The sum of all the elements is: ", sum)
```

### Assignment 12: Factorial Calculation

Write a program that calculates the factorial of a number input by the user using a while loop.

```python
num = int(input("Write a number for calculating his factorial"))

res = 1

for i in range(2, num + 1):
    res *= i

print("Factorial of ", num, " is", res)    
```

### Assignment 8: break Statement

Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

```python
sum = 0

num = int(input("Please write a number, it is going to add to the sum of numbers if you want to stop adding numbers presss 0"))

while num != 0:
    sum += num
    num = int(input("Please write a number, it is going to add to the sum of numbers if you want to stop adding numbers presss 0"))

print("The sum is: ", sum)
```

### Assignment 15: Fibonacci Sequence

Write a program that prints the first n Fibonacci numbers, where n is input by the user.

```python
i = 0
actual = 0
before = 0
fib = 0

n = int(input("Select which Fibonacci number you want to get"))

while i < n:
    if i == 0:
        print(0, end = " ")
        actual = 1
    elif i == 1:
        print(actual, end = " ")
    else:
        fib = actual + before
        before = actual
        actual = fib
        print(fib, end = " ")
    i = i + 1
```
