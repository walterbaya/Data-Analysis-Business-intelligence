# Pandas-Dataframe y Series

Pandas es una librería de manipulacion de datos en Python, que se usa para el analisis y limpieza dispone de dos estructuras, una son las series y la otra los dataframes

```python
import pandas as pd
```

```python
#Series
#Son arreglos que pueden tener cualquier tipo de datos, las series son de una dimension

data=[1,2,3,4,5]

series = pd.Series(data)
print("Series \n", series)
```

```python
#Crear series desde un diccionario

data = {'a': 1, 'b': 2, 'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data=[10, 20, 30]
index = ['a', 'b', 'c']
pd.Series(data, index=index)
```

```python
#Dataframes

#Crear un Dataframe desde un diccionario

data={
    'Name': ['Krish', 'John', 'Jack'],
    'Age': [25, 30 ,45],
    'City': ['Bangalore', 'New York', 'Florida']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np

np.array(df)
```

```python
## Crear un Data Frame a partir de una lista de diccionarios

data = [
    {'Name': 'Krish', 'Age': 32, 'City': 'Bangalore'}, 
    {'Name': 'John', 'Age': 34, 'City': 'New York'},
    {'Name': 'Bappy', 'Age': 32, 'City': 'Bangalore'},
    {'Name': 'Jack', 'Age': 32, 'City': 'Bangalore'}  
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
