```python
import seaborn as sns
```

```python
#Plots basicos

tips = sns.load_dataset('tips') #Cargamos un dataset de los por defecto
tips
```

```python
#Crear un scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tips")
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line plot of Total bill by size")
plt.show()
```

```python
# Categorical Plots 
# Bar plot
sns.barplot(x ='day', y='total_bill', data=tips)
plt.title('Bar Plot para total bill')
```

```python
# Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
#Violin plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# Histogramas

sns.histplot(tips['total_bill'], bins=10, kde=True)

#kde te pone la curva
```

```python
## KDE plot

sns.kdeplot(tips['total_bill'], shade=True)
```

```python
# Pairplot 
# Muestra la relacion entre variables

sns.pairplot(tips)
```

```python
#Heatmap

corr = tips[['total_bill', 'tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
